```python
# remove outliers for multiple columns
def cap_data(data_new):
    for col in data_new.columns:
        print("capping the ",col)
        if (((data_new[col].dtype)=='float64') | ((data_new[col].dtype)=='int64')):
            percentiles = data_new[col].quantile([0.01,0.99]).values
            data_new[col][data_new[col] <= percentiles[0]] = percentiles[0]
            data_new[col][data_new[col] >= percentiles[1]] = percentiles[1]
        else:
            data_new[col]=data_new[col]
    return data_new

final_data_new=cap_data(data_new)
```
